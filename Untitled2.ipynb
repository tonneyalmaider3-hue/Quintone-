```python
import numpy as np

# Set a seed for reproducibility of the random data
np.random.seed(42)

# 1. Create a 2D array representing transaction volumes
# [span_0](start_span)4 branches (rows) over 6 months (columns)[span_0](end_span)
# Using random integers between 1000 and 10000 for realistic transaction volumes
transaction_volumes = np.random.randint(1000, 10000, size=(4, 6))

print("1. 4x6 Transaction Volumes Array (Rows=Branches, Columns=Months):\n", transaction_volumes)
# --------------------------------------------------

# 2. Calculate the total transactions per branch
# [span_1](start_span)Summing across the columns (axis=1)[span_1](end_span)
total_transactions_per_branch = transaction_volumes.sum(axis=1)

print("\n2. Total Transactions Per Branch (Branch 1 to 4):\n", total_transactions_per_branch)
# --------------------------------------------------

# 3. Identify the branch with the highest total transactions
# [span_2](start_span)Find the index of the maximum value[span_2](end_span)
highest_branch_index = total_transactions_per_branch.argmax()

# The branch number is the index + 1
branch_number = highest_branch_index + 1
highest_volume = total_transactions_per_branch[highest_branch_index]

print(f"\n3. Branch with the Highest Total Transactions: Branch {branch_number}")
print(f"   Highest Total Volume: {highest_volume}")
# --------------------------------------------------

# 4. Compute the average monthly transaction volume across all branches
# [span_3](start_span)Calculate the mean of the entire 2D array[span_3](end_span)
average_monthly_volume = transaction_volumes.mean()

print("\n4. Average Monthly Transaction Volume Across All Branches:")
print(f"   {average_monthly_volume:.2f}")
# --------------------------------------------------

# 5. Reshape the array to a 3x8 format and explain the implications
# The original array has 4 * 6 = 24 elements.
# [span_4](start_span)Reshaping to 3x8 is valid since 3 * 8 = 24[span_4](end_span).
try:
    reshaped_array = transaction_volumes.reshape(3, 8)
    print("\n5. Reshaped Array (3x8):\n", reshaped_array)

except ValueError as e:
    print(f"\n5. Reshaping Error: {e}")
    print("   Reshaping to 3x8 is only possible if the total number of elements (24) is unchanged.")

# [span_5](start_span)Explanation of implications[span_5](end_span)
print("\n   Implications of Reshaping from 4x6 to 3x8:")
print("   * **Data Integrity:** The total number of elements remains **24**, so the underlying data is preserved.")
print("   * **Context Loss:** The original meaning (4 branches x 6 months) is **lost**.")
print("   * **New Interpretation:** The new $3 \\times 8$ array would now represent **3 entities** (e.g., regions or segments) over **8 new periods** (e.g., fortnights) or simply a different logical grouping, which is not aligned with the original 4-branch, 6-month financial context.")
print("   * **Use Case:** Reshaping is often used for operations like feeding data into certain machine learning models or for display purposes, where the original 'branch' and 'month' context is less critical than the matrix dimensions themselves.")
```

```python
# 1. Create a list of 10 delivery routes, each represented as a string.
delivery_routes = [
    "North Industrial Park",
    "Central Business District",
    "East Residential Area",
    "West Farm Belt",
    "Newtown Suburbs",
    "Old Port Authority",
    "National Highway 1 Run",
    "Midtown Retail Zone",
    "Airport Cargo Loop",
    "Northern Valley Route"
]
print("1. Initial List of 10 Delivery Routes:")
print(delivery_routes)

# --------------------------------------------------

# 2. Append a new route and remove a discontinued one.
# Append a new route
delivery_routes.append("South Coastline Express")

# Remove a discontinued route (e.g., "West Farm Belt")
# Using .remove() for a specific value
try:
    delivery_routes.remove("West Farm Belt")
except ValueError:
    print("\nError: 'West Farm Belt' not found in the list.")

print("\n2. List after Appending 'South Coastline Express' and Removing 'West Farm Belt':")
print(delivery_routes)

# --------------------------------------------------

# 3. Sort the list alphabetically and reverse it.
# Sort alphabetically
delivery_routes.sort()
print("\n3a. List Sorted Alphabetically:")
print(delivery_routes)

# Reverse the sorted list
delivery_routes.reverse()
print("\n3b. List Reversed:")
print(delivery_routes)

# --------------------------------------------------

# 4. Count how many routes start with the letter "N".
n_route_count = 0
for route in delivery_routes:
    # Use .startswith() method
    if route.startswith("N"):
        n_route_count += 1

# Alternative using a generator expression and sum:
# n_route_count = sum(1 for route in delivery_routes if route.startswith("N"))

print('\n4. Count of Routes Starting with the letter "N":')
print(f"There are {n_route_count} routes starting with 'N'.")

# --------------------------------------------------

# 5. Use list comprehension to create a new list of routes longer than 10 characters.
# The condition is len(route) > 10
long_routes = [route for route in delivery_routes if len(route) > 10]

print("\n5. New List of Routes Longer Than 10 Characters (using list comprehension):")
print(long_routes)
```

```python

# Create a dictionary for product inventory
inventory = {
    "Laptop Pro X": 45,
    "Wireless Mouse Z": 150,
    "4K Monitor 27": 8,
    "Ergo Keyboard V3": 95,
    "Webcam HD 500": 3
}

print("1. Initial Inventory Dictionary:")
print(inventory)

# Add a new product
inventory["USB-C Hub 8-in-1"] = 65

# Update the quantity of an existing product ("Laptop Pro X" received 10 more units)
inventory["Laptop Pro X"] += 10 # 45 + 10 = 55

print("\n2. Inventory after Adding a New Product and Updating Stock:")
print(inventory)

def get_low_stock_products(inventory_dict):
    """Returns a new dictionary containing only products with stock < 10."""
    low_stock = {}

    # Iterate through key-value pairs (product, quantity)
    for product, quantity in inventory_dict.items():
        if quantity < 10:
            low_stock[product] = quantity

    return low_stock

# Call the function
low_stock_items = get_low_stock_products(inventory)

print("\n3. Low Stock Products (Quantity < 10):")
print(low_stock_items)





# Identify a discontinued product (e.g., "Webcam HD 500")
discontinued_product = "Webcam HD 500"

# Delete the product
if discontinued_product in inventory:
    del inventory[discontinued_product]

print(f"\n4. Inventory after Deleting Discontinued Product ('{discontinued_product}'):")
print(inventory)
print("\n5. Looping through Inventory using .items():")
for product, quantity in inventory.items():
    print(f"   Product: {product}, Stock Quantity: {quantity}")
```
